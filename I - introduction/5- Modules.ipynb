La véritable force de Python repose sur sa communauté. Des développeurs ont créé des modules que l'ont peut réutiliser. Les répertoires de dépôts les plus connus sont  https://pypi.org/ et https://github.com/.

### Les packages

Les packages sont une collection de modules organisés en répertoires en sous-répertoires qui détiennent un fichier `__init__.py`.

On a créé un package dans ce dossier **first_package**.

#### Installer un package
On peut installer un package depuis Jupyter Notebook avec la commande :

%pip install

Par exemple :

%pip install mstarpy

Ou depuis le terminal avec :

pip install mstarpy

NB : pour instaler une version spécifique on utiliser pip install mstarpy==2.0.0

#### Mettre à jour un package

Une fois un package installer, on peut le mettre à jour pour avoir la verison la plus récente.

%pip install mstarpy --upgrade

Ou depuis le terminal avec :

pip install mstarpy --upgrade


#### Désinstaller un package
Pour désinstaller un package 
%pip uninstall mstarpy -y

Ou depuis le terminal avec :

pip uninstall mstarpy -y


#### Utiliser un package

Un fois installer, on peut utiliser un package en l'important.

```python
import mstarpy
```

NB : On peut aussi donner un alias au package qui sera utilisé pour l'appeler.

import mstarpy as ms
