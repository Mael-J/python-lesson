## Transformer le fichier en DataFrame

```python
# transformer le fichier .csv en DataFrame, séprarateur ; et décimal ,
df = pd.read_csv('country_gdp_per_capita.csv', sep=';', decimal=',')
```

## Transposer

La méthode `transpose` permet d'inverser les indices et les colonnes. Les indices deviennent des colonnes, les colonnes deviennent des indices.

La propriété `T` réalise la même opération.

```python
#transpose le DataFrame avec la méthode transpose
df.transpose()
```

```python
#transpose le DataFrame avec l'attribut T
df.T
```

## stack

Notre objectif est de transformer notre DataFrame dont les colonnes sont les années en un DataFrame qui comporte 4 colonnes le code pays, le nom du pays, l'année et le niveau du PIBs.

Allons-y étapes par étapes.

### Etape 1 : définir l'indice

Notre indice sera le code pays et le nom du pays

```python
#indice code et nom du pays
df.set_index(["Country Code", "Country Name"])
```

### Etape 2 : orienter les colonnes en lignes

Les années en colonne deviennent des lignes ave cla méthode `stack`.

```python
#orienter les colonnes en ligne
df.set_index(["Country Code", "Country Name"]).stack()
```

### Etape 3 : recréer un indice

La méthode `reset_index` permet de récréer un indice et de transformer l'indice composé de `Country Code` et `Country Name` en colonnes.

```python
#transforme l'index en colonnes et redéfini l'index
df.set_index(["Country Code", "Country Name"]).stack().reset_index()
```

### Etape 4 : renommer les colonnes

Le DataFrame précédent a la forme finale souhaitée, cependant ces noms de colonnes ne sont pas explicites, nous allons les renommer avec la méthode `rename`.

```python
#renomme les colonnes
df.set_index(["Country Code", "Country Name"]).stack().reset_index().rename(columns={"level_2": "Year", 0: "GDP"})
```

### Etape 5 : conserver le nouveau DataFrame

Nous avons finalement créé notre DataFrame, nous allons conservé son résultat dans une nouvelle variable pour l'utiliser dans la suite du cours.

```python
#nouveau DataFrame
df_stack = df.set_index(["Country Code", "Country Name"]).stack().reset_index().rename(columns={"level_2": "Year", 0: "GDP"})
```

```python
df_stack.groupby([""])
```

```python
df_pivot = df_stack.pivot(index = "Country Code", columns = "Year", values = "GDP").reset_index()
```
