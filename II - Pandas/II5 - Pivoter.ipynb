# Faire pivoter le DataFrame

Pour réaliser certaines opérations et des calculs, notre DataFrame doit avoir une forme spécifique, c'est pourquoi on doit le transformer.

Dans ce cours, nous allons voir comment pivoter le DataFrame et changer l'orientation des lignes et des colonnes.

## importer pandas

```python
#import pandas
import pandas as pd
```

## Transformer le fichier en DataFrame

```python
# transformer le fichier .csv en DataFrame, séprarateur ; et décimal ,
df = pd.read_csv('country_gdp_per_capita.csv', sep=';', decimal=',')
```

```python
#aprcçu du fichier
df.head(5)
```

## Transposer

La méthode `transpose` permet d'inverser les indices et les colonnes. Les indices deviennent des colonnes, les colonnes deviennent des indices.

La propriété `T` réalise la même opération.

```python
#transpose le DataFrame avec la méthode transpose
df.transpose()
```

```python
#transpose le DataFrame avec l'attribut T
df.T
```

## stack

Notre objectif est de transformer notre DataFrame dont les colonnes sont les années en un DataFrame qui comporte 4 colonnes le code pays, le nom du pays, l'année et le niveau du PIBs.

Allons-y étapes par étapes.

### Etape 1 : définir l'indice

Notre indice sera le code pays et le nom du pays

```python
#indice code et nom du pays
df.set_index(["Country Code", "Country Name"])
```

### Etape 2 : orienter les colonnes en lignes

Les années en colonne deviennent des lignes ave cla méthode `stack`.

```python
#orienter les colonnes en ligne
df.set_index(["Country Code", "Country Name"]).stack()
```

### Etape 3 : recréer un indice

La méthode `reset_index` permet de récréer un indice et de transformer l'indice composé de `Country Code` et `Country Name` en colonnes.

```python
#transforme l'index en colonnes et redéfini l'index
df.set_index(["Country Code", "Country Name"]).stack().reset_index()
```

### Etape 4 : renommer les colonnes

Le DataFrame précédent a la forme finale souhaitée, cependant ces noms de colonnes ne sont pas explicites, nous allons les renommer avec la méthode `rename`.

```python
#renomme les colonnes
df.set_index(["Country Code", "Country Name"]).stack().reset_index().rename(columns={"level_2": "Year", 0: "GDP"})
```

### Etape 5 : conserver le nouveau DataFrame

Nous avons finalement créé notre DataFrame, nous allons conservé son résultat dans une nouvelle variable pour l'utiliser dans la suite du cours.

```python
#nouveau DataFrame
df_stack = df.set_index(["Country Code", "Country Name"]).stack().reset_index().rename(columns={"level_2": "Year", 0: "GDP"})
```

### Calculer l'évolution du PIB par habitant par pays

Le DataFrame `df_stack` a la forme qui convient pour calculer l'évolution par pays du PIB par habitant.

Les années de la colonne `Year` sont une chaîne de texte. Nous les transformons en nombre entier pour ensuite les trier de la date la plus ancienne à la plus récente.

```python
# les années sont au format texte
df_stack.dtypes
```

```python
#transformer les années en nombre entier
df_stack["Year"] = df_stack["Year"].astype('int')
#les années sont au format entier
df_stack.dtypes
```

```python
# on trie les années par date croissante
df_stack.sort_values("Year",ascending =True)
```

```python
# on calculer l'évolution du PIB par pays
df_stack.sort_values("Year",ascending =True).groupby(["Country Code"])["GDP"].pct_change()
```

```python
#on conserve le résultat dans une nouvelle colonne
df_stack["evolution GDP"] = df_stack.sort_values("Year",ascending =True).groupby(["Country Code"])["GDP"].pct_change()
```

```python
#aperçu du DataFrame avec la nouvelle colonne evolution GDP
df_stack.head(5)
```

## Pivoter le DataFrame

Nous pouvons transformer notre DataFrame `df_stack` pour obtenir les années en colonnes et code pays en indice avec la méthode `pivot`.

```python
#pivote le DataFrame
df_stack.pivot(index = "Country Code", columns = "Year", values = "evolution GDP")
```

## Concatener

Plusieurs DataFrame peuvent être concaténés pour former un grand DataFrame.

Reprenons notre DataFrame `df_stack` et créons trois nouveaux DataFrames avec trois pays différents.

```python
# DataFrame 1 : France
df_france = df_stack.loc[df_stack["Country Code"] == 'FRA']
# aperçu du DataFrame
df_france.head(5)
```

```python
# DataFrame 2 : Thailande
df_thailand = df_stack.loc[df_stack["Country Code"] == 'THA']
# aperçu du DataFrame
df_thailand.head(5)
```

```python
# DataFrame 3 : Chili
df_chile = df_stack.loc[df_stack["Country Code"] == 'CHL']
# aperçu du DataFrame
df_chile.head(5)
```

L'argument axis = 0 superpose les DataFrames suivant les colonnes. L'argument ignore_index = True redéfinit un indice.

```python
#superpose les DataFrames
pd.concat([df_france, df_thailand, df_chile], axis=0, ignore_index=True)
```

L'argument axis = 1 concatène les DataFrames suivant les indices, l'indice 0 est associé à l'indice 0, l'indice 42 à l'indice 42... C'est pourquoi il est important de redéfinir les indices lors de la création de nos DataFrames.

```python
# rédéfini les indices du DataFrame France
df_france = df_france.reset_index(drop=True)
# rédéfini les indices du DataFrame Thailande
df_thailand = df_thailand.reset_index(drop=True)
# rédéfini les indices deu DataFrame Chili
df_chile = df_chile.reset_index(drop=True)
```

```python
pd.concat([df_france, df_thailand, df_chile], axis=1)
```
